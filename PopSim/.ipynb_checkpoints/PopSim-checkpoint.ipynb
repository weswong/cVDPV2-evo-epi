```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
import json
from collections import defaultdict, Counter
```

```python
class Individual:
    id_generator = itertools.count()
    scale = 1.0
    def __init__(self, age, scale = 1): #fertility_scale is one of the arguments of the script
        self.id = next(Individual.id_generator) #iterator that generates a new unique id everytime it is called
        self.age = age
        self.isalive =True
        self.check_fertility()
        
        
    def check_alive(self, dt): #check if person is still alive after specified time period
        p_death = 1-np.e**(-self.mortality_rate()*dt)
        if np.random.random() < p_death:
            self.isalive = False
            self.death_age = self.age
            
    def check_fertility(self):
        if self.age >=15 and self.age <50: #assume males virility follow fertile rules in females
            self.isfertile = True
        else:
            self.isfertile = False
    
    @classmethod
    def fertility_rate_fn(cls, age, a=-0.012187592249717867, b=0.6277666066187078, c=-9.858673931204683):
            if age < 15 or age > 50:
                return 0
            else:
                return np.e**(a * age**2 + b* age + c) *Individual.scale
    @classmethod
    def p_child_fn(cls, age, dt):
        f_rate = Individual.fertility_rate_fn(age)
        p_child =1-np.e**(-f_rate*dt)
        return p_child 

    @classmethod
    def create_child(cls):
        if np.random.random() > 0.5:
            return Male(0)
        else: 
            return Female(0)


class Male(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.sex = 1
    
    def mortality_rate(self, minimum_mortality = 0.001):
        #the age thresholds for child_fn and elder_fn are guestimated from the data -- will need to check this when applying to other countries
        #minimum mortality is also guestimated from the data -- this is mostly to do away with a 0 mortality rate
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.038012826363843 + -0.4620757529135555*self.age))
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(-9.57843000677779 + 0.08987439511446728*self.age))
        else:
            m = minimum_mortality
        return m
    
    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:
            self.age += dt
            self.check_fertility()
    

class Female(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.n_children = 0 #keep track of n_children only on female side
        self.sex =0
        self.refractory_period = 0 #period of time where she does not give birth again -- rounding it to 1 year-- this could probably be turned off tbh
    
    def mortality_rate(self, minimum_mortality = 0.001):
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.154414473758413 + -0.4420609900131315*self.age)) #m = age specific mortality rate (deaths/individual/year)
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(-10.527925730119318 + 0.10104736306154925*self.age))
        else: 
            m= minimum_mortality
        return m
    
    def birth(self, dt):
        '''determines whether a woman gives birth. 
        other than imposing a 1 year refractory period births are 
        treated as independent of one another.. may want to change this'''
        
        if self.isfertile and self.refractory_period <=0:
            p_birth = Individual.p_child_fn(self.age,dt)
            
            if np.random.random() <= p_birth:
                self.n_children +=1
                self.refractory_period = 1 #impose a one year refractory period, but otherwise births
                return True
            else:
                return False
        else: 
            return False
                
    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:            
            self.age += dt #advance time
            self.check_fertility() #advance time            
            if self.refractory_period > 0:
                self.refractory_period -= dt #going into negative is OK
```

```python
class Household:
    '''Household defined by active couple units that generate children
    Allow for multi-generational living (grandparents, children, and grandchildren cohabitating)
    Centered around the concept of a family core that represents the unit that is capable of producing children'''
    id_generator = itertools.count()
    def __init__(self):
        self.core = []
        self.children_id = []
        self.individuals = {} # a container to hold the individual classes relevant to this household
        self.id = next(Household.id_generator)

    @classmethod
    def initialize_household(cls, founder_list, children_list = None): #going to assume I1 and I2 are always opposite sex
        HH = Household()
        for individual in founder_list:
            HH.individuals[individual.id] = individual
            HH.core.append(individual.id)
        if children_list:
            for child in children_list:
                HH.individuals[child.id] = child
                HH.children_id.append(child.id)
        HH.check_alive()
        HH.check_core_status()
        return HH
            
    def check_core_status(self):
        '''active means that the core can produce children
        receptive means that the core is willing to find a mate
        receptive = 1 member in core who is fertile
        This must be run immediately after the core is modified, for any reason'''
        
        #if statements ~2x slower
        if len(self.core) == 2:
            self.core_receptive = None
            iid1, iid2 = self.core[0], self.core[1]
            if self.individuals[iid1].isfertile and self.individuals[iid2].isfertile:
                if self.individuals[iid1].sex != self.individuals[iid2].sex:
                    self.core_active = True
                else:
                    raise ValueError('Warning Same Sex in {hid}'.format(hid = self.id))
                    self.core_active = False
        
        elif len(self.core) == 1:
            iid = self.core[0]
            self.core_active = False
            if self.individuals[iid].isfertile: 
                self.core_receptive = 1. - self.individuals[iid].sex #if the person in the core is fertile, the core is receptive of a new mate
            else:
                self.core_receptive = None 
        else:
            self.core_active, self.core_receptive = False, None
            
            
    def remove_member(self, iid, keep_individual = False):
        if iid in self.children_id:
            self.children_id.remove(iid)
        if iid in self.core: 
            #if one member of the core is removed, the entire core will be replaced by the eldest available child
            #assumes mating for life
            self.core.remove(iid)
            eldest_child = self.find_eldest_child() #can return None
            if eldest_child:
                self.core = [eldest_child]
            self.check_core_status()
                
        if keep_individual == False:
            self.individuals.pop(iid) #immediately discard
        else:
            return self.individuals.pop(iid)
        self.check_alive()
                
    def remove_dead(self):
        dead_ids = [iid for iid in self.individuals if self.individuals[iid].isalive == False]
        for iid in dead_ids:
            self.remove_member(iid)
            
    def remove_adult_children(self, move_out_age = 18):
        '''children who succeed the family are automatically removed from children list during succession'''
        adult_children = [self.remove_member(iid, True) for iid in self.children_id if self.individuals[iid].age >= move_out_age]
        self.check_alive()
        return adult_children
    
    def check_alive(self):
        if len(self.individuals) >=1:
            self.alive = True #ie is the Household alive
        else:
            self.alive = False
    
    def add_core(self, individual):
        self.individuals[individual.id] = individual
        self.core.append(individual.id)
        self.check_core_status()
        
    def attempt_child(self, dt):
        core_f_id= [iid for iid in self.core if self.individuals[iid].sex == 0][0] #f_id = female id

        if self.individuals[core_f_id].birth(dt):
            child = Individual.create_child()
            self.individuals[child.id] = child
            self.children_id.append(child.id)
        
    def find_eldest_child(self):
        try:
            eldest_child_id = min(self.children_id) #ids are generated sequentially, smallest index = child been around longer
        except:
            eldest_child_id = None
        return eldest_child_id   
    
    def attempt_succession(self):
        '''succession occurs when the family core is either 
        too old or if one of the members of the original core dies'''
        eldest_child = self.find_eldest_child()
        if eldest_child: 
            self.core = [eldest_child]
            self.children_id.remove(eldest_child)
            #print('succession occured {hid} {id}'.format(hid = self.id, id = eldest_child))
        else:
            self.core = []
        self.check_core_status()
        self.check_alive()

    
    @classmethod
    def combine_household(cls, hh1, hh2):
        '''merge the core of hh2 into hh1 -- only do for single households!'''
        hh2_iid = hh2.core.pop(0) #it should be the only thing in that list
        hh2_individual = hh2.individuals.pop(hh2_iid)
        hh2.check_core_status()
        hh2.check_alive()
        
        hh1.add_core(hh2_individual)
        hh1.check_core_status()
        hh1.check_alive()
        
        
                    
    def update(self, dt=1, remove_adult_children = False):
        for iid in self.individuals:
            self.individuals[iid].update(dt)
        self.remove_dead()
        self.check_alive()
        self.check_core_status()
        
        if self.alive:
            if self.core_active:
                self.attempt_child(dt)
            
            elif self.core_receptive == None: #ie the core is not active and no longer receptive
                self.attempt_succession()
            
            if remove_adult_children:
                return self.remove_adult_children() #removing adult children not usually done at this step. Usually done at the Population level since it needs the adult children
                    
                    
    def return_household_size(self):
        return len(self.individuals)
    
    def return_children_size(self):
        return len(self.children_id)
        
        
```

```python
def run(iterations = 1000, remove_adult_children = True):
    sims = []
    for _ in range(iterations):
        household_size = []
        hh = Household.initialize_household([Male(18), Female(18)])
        for t in range(150):
            hh.update(remove_adult_children = remove_adult_children)
            household_size.append(hh.return_household_size())
        sims.append(household_size)
    return np.median(sims, axis =0), np.percentile(sims, 2.5, axis =0), np.percentile(sims, 97.5, axis = 0)

sims_adults_removed = run(10000)
sims_adults_stay = run(10000, False)
```

Single household dynamics over time.

```python
class Population:
    def __init__(self):
        self.households = {} #households that are "receptive" of either a fertile female (0) or male (1)        
        self.receptive_households = {1:[], 0:[]}
        
    @classmethod
    def initialize_population(cls, n_households= 100):
        P = Population()
        for _ in range(n_households):
            hh = Household.initialize_household([Male(18), Female(18)])
            P.households[hh.id] = hh
        return P
    
    def prime(self):
        '''identify the adult children and put them into a singles pool
        identify receptive households (categorized by the sex they need)'''
        singles_pool = []
        receptive_households = {1:[], 0:[]}
        
        for hid in self.households:
            singles_pool += self.households[hid].remove_adult_children()
            
            if self.households[hid].core_receptive == 1: #hh needs male
                receptive_households[1].append(hid)
                
            elif self.households[hid].core_receptive ==0: #hh needs female
                receptive_households[0].append(hid)
        
        np.random.shuffle(receptive_households[0]) #shuffle the lists in place so I don't end up putting kids back to their original home
        np.random.shuffle(receptive_households[1])
        
        #Take any existing receptive households and try to pair them
        leftover_receptive_households = {1:[], 0:[]}
        for f,m in itertools.zip_longest(receptive_households[0], receptive_households[1]):
            if f == None:
                leftover_receptive_households[1].append(m)
            elif m == None:
                leftover_receptive_households[0].append(f)
            else:
                #print('{f} ({f_sex}) and {m} ({m_sex}) combined'.format(f=f, m=m, f_sex=str(self.households[f].core_receptive), m_sex = str(self.households[m].core_receptive)))               
                
                #if len(self.households[m].core) < 1:
                #    print(self.households[m].__dict__)
                
                Household.combine_household(self.households[f],self.households[m])
                self.households.pop(m)
        return singles_pool, leftover_receptive_households
    
    def distribute_singles(self):
        singles_pool, receptive_households = self.prime()
        waiting_pool ={0:[], 1:[]} #individuals who cannot find a receptive hh
        
        while len(singles_pool) > 0:
            single = singles_pool.pop(0) #iteratively remove the first element of the list over and over and over
            
            #first assign to receptive households
            if len(receptive_households[single.sex]) > 0:
                assigned_hid = receptive_households[single.sex].pop(0)
                #print('{id1} ({single_sex}) assigned to household {household_id} needing ({missing_sex})'.format(id1 = single.id, single_sex = single.sex, household_id = assigned_hid, missing_sex = self.households[assigned_hid].core_receptive))
                self.households[assigned_hid].add_core(single)
                
                
            else: #if no hh are available for that particular sex, throw them into a waiting pool
                waiting_pool[single.sex].append(single)
        
        #second: pair remaining individuals in the waiting pool
        zipped = itertools.zip_longest(waiting_pool[0], waiting_pool[1]) #different from a regular zip, it will pad the shorter list with "None" values
        for f,m in zipped:
            core = [f,m]
            try:
                core.remove(None) #could not find mate within the waiting pool
                #print('new single hh created {id}'.format(id = core[0].id))
            except:
                #print('new hh created {f_id} ({f}) and {m_id} ({m})'.format(f_id = f.id, m_id = m.id, f = f.sex, m = m.sex))
                pass
            hh = Household.initialize_household(core)
            self.households[hh.id] = hh
        
    def remove_dead(self):
        dead_ids = [hid for hid in self.households if self.households[hid].alive == False]
        for hid in dead_ids:
            self.households.pop(hid)
            #print('removing dead hh')
            
    def update(self,dt=1):
        for hid in self.households:
            self.households[hid].update(dt)
        self.remove_dead()
        self.distribute_singles()
        self.remove_dead() 
    
    def return_household_size(self):
        return Counter([hh.return_household_size() for hh in self.households.values()])
    
    def return_children_size(self):
        return Counter([hh.return_children_size() for hh in self.households.values() if hh.return_children_size() != 0])
    
    def return_age_pyramid(self,sex):        
        age_pyramid = [0 for _ in range(0,105,5)]
        ages = []
        for hh in self.households.values():
            ages += [individual.age for individual in hh.individuals.values() if individual.sex == sex]
            
        bins = [x for x in range(0, 105,5)] # agebins
        bin_data = np.digitize(ages, bins)
        for idx in bin_data:
            age_pyramid[idx-1] += 1
        return age_pyramid   
    
    def return_TFR(self):
        """Return the total fertility rate (n_children per woman past child birth age (50) )"""
        tfr = []
        for hh in self.households.values():
            tfr += [individual.n_children for individual in hh.individuals.values() if (individual.sex == 0) and (individual.age > 50)]
        if len(tfr) == 0:
            tfr = [0]
        return tfr
    
    def return_n_individuals(self):
        n_individuals = [0,0]
        for hh in self.households.values():
            for i in hh.individuals.values():
                n_individuals[i.sex] += 1
        return n_individuals 
            
        
            
                
```

```python
class Simulation:
    def __init__(self, n_households, n_tsteps, dt, fertiliy_scale):
        self.params = {'n_tsteps': n_tsteps,
                       'dt': dt,
                       'n_households': n_households, 
                       'fertility_scale': fertility_scale}
        self.age_pyramid = {0:[], 1:[]}
        self.household_size = []
        self.children_size = []
        self.tfr = []
        self.n_individuals = {0:[], 1:[]}

        #self.P = None
        
    @classmethod
    def run(cls, n_households=5000, n_tsteps=500, dt=1, fertility_scale = 1):
        Individual.scale = fertility_scale
        S = Simulation(n_households, n_tsteps, dt, fertility_scale)
        P = Population.initialize_population(n_households)
        S.add_data(P)
        
        for _ in range(n_tsteps):
            P.update(dt)
            S.add_data(P)
    
        return S
    def add_data(self, P):
        self.age_pyramid[0].append(P.return_age_pyramid(0))
        self.age_pyramid[1].append(P.return_age_pyramid(1))
        self.household_size.append(P.return_household_size())
        self.children_size.append(P.return_children_size())
        self.tfr.append(P.return_TFR())
        
        n_individuals = P.return_n_individuals()
        self.n_individuals[0].append(n_individuals[0])
        self.n_individuals[1].append(n_individuals[1])
    def write(self, out_file='sims/results.json'):
        with open(out_file, 'w') as fp:
            json.dump(self.__dict__, fp)
            
```
