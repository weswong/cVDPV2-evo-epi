```python
def family_type(self):
        #find the max distance between LIVING nodes
        '''find the max distance between the eldest node and the terminal living nodes (leaf)
        nuclear: distance <= 1
        joint/multi-generational: distance > 1'''
        leaves = self.tree.get_leaves()
        eldest_node = self.find_eldest_node()
        distances = [self.tree.get_distance(eldest_node, leaf.name) for leaf in leaves if self.tree_map[leaf.name].status != 'dead' and eldest_node != leaf.name]
        
        if len(self.tree_map) > 1 and len(distances) > 0:
            if np.max(distances) == 1:
                if len([n.status for n in self.tree_map.values() if n.status == 'married']) == 1:
                    return 'strict-nuclear'
                else:
                    return 'extended-nuclear'
            elif np.max(distances) > 1:
                return 'multi-generational'
        elif len(self.tree_map) == 1:
            return self.tree_map[eldest_node].status
        else:
            return 'misc'
        
hh_age_distribution = defaultdict(list)
for v in V_immediate.villages:
    for bari in v.units.values():
        for hh in bari.households.values():
            for individual in hh.individuals.values():
                if individual.age <=1:
                    hh_age_distribution[hh.n_individuals()].append('<1')
                elif individual.age < 5:
                    hh_age_distribution[hh.n_individuals()].append('1-5')
                elif individual.age <15: 
                    hh_age_distribution[hh.n_individuals()].append('5-15')
                elif individual.age < 60:
                    hh_age_distribution[hh.n_individuals()].append('15-60')
                else:
                    hh_age_distribution[hh.n_individuals()].append('>60')
            
for key in hh_age_distribution:
    hh_age_distribution[key] = Counter(hh_age_distribution[key])
max_size = np.max(list(hh_age_distribution.keys()))
infants = [hh_age_distribution[key]['<1'] if key in hh_age_distribution.keys() and '<1' in hh_age_distribution[key].keys() else 0 for key in range(1, max_size)]
young_children = [hh_age_distribution[key]['1-5'] if key in hh_age_distribution.keys() and  '1-5' in hh_age_distribution[key].keys() else 0 for key in range(1, max_size)]
children = [hh_age_distribution[key]['5-15'] if key in hh_age_distribution.keys() and '5-15' in hh_age_distribution[key].keys() else 0 for key in range(1, max_size)]
adults = [hh_age_distribution[key]['15-60'] if key in hh_age_distribution.keys() and '15-60' in hh_age_distribution[key].keys() else 0 for key in range(1, max_size)]
elders = [hh_age_distribution[key]['>60'] if key in hh_age_distribution.keys() and '>60' in hh_age_distribution[key].keys() else 0 for key in range(1, max_size)]

import seaborn as sns
cpalette = sns.color_palette("Oranges", 5)
y = np.asarray([infants, young_children, children, adults, elders])
# this call to 'cumsum' (cumulative sum), passing in your y data, 
# is necessary to avoid having to manually order the datasets
x_axis = range(1,np.max(list(hh_age_distribution.keys()))) 

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.stackplot(x_axis, infants, young_children, children, adults, elders, labels = ['<1','1-5', '5-15', '15-60', '60+'], colors = cpalette)
plt.xticks(x_axis, x_axis)
plt.title('Household Age Composition', fontsize = 20)
plt.ylabel('Frequency', fontsize = 15)
plt.xlabel('hh size', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

prop_y = []
for i,data in enumerate(y.T):
    prop_y.append(data / np.sum(y.T, axis = 1)[i])
prop_y = np.asarray(prop_y).T

plt.subplot(1,2,2)
plt.stackplot(x_axis, prop_y, labels = ['<1','1-5', '5-15', '15-60', '60+'], colors = cpalette)
plt.xticks(x_axis, x_axis)
lgd = plt.legend(loc=(1.05, 0.3), fontsize = 15)
plt.title('Household Age Composition', fontsize = 20)
plt.ylabel('Proportion', fontsize = 15)
plt.xlabel('hh size', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.savefig('household_age_composition.svg', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
def prepare_contact_data(V_test_object):    
    sim_results = np.zeros((len(V_test_object.villages), 100))
    for vidx,V in enumerate(V_test_object.villages):
        contact_ages = []
        infants,young_children = [], []
        for bari in V.units.values():
            for hh in bari.households.values():
                for individual in hh.individuals.values():
                    if individual.age <=1:
                        infants.append((vidx, bari.id, hh.id, individual.id))
        for infant in infants:
            vid, bid, hhid, iid = infant

            hh = V_test_object.villages[vid].units[bid].households[hhid]
            individual_ages = []
            for individual in hh.individuals.values():
                if individual.id != iid:
                    if individual.age <= 14:
                        individual_ages.append(individual.age)
                    else:
                        if individual.sex == 0:
                            individual_ages.append(individual.age)

                    #elif individual.sex == 0:
                    #    individual_ages.append(individual.age)
            #individual_ages = sorted([individual.age for individual in hh.individuals.values() if individual.id != iid  and individual.sex == 0])
            if len(individual_ages) > 2:
                youngest_two = sorted(individual_ages)[0:2]
            else:
                youngest_two = sorted(individual_ages)
            contact_ages += youngest_two
        contact_age_counts = Counter(contact_ages)
        total = np.sum(list(contact_age_counts.values()))
        y_axis = [contact_age_counts[_]/total if _ in contact_age_counts.keys() else 0 for _ in range(100)]
        sim_results[vidx] = y_axis
    return sim_results


def plot_contact_age_distribution(V_test_object,label, color):
    plt.figure(figsize=(10,5))
    sim_results = prepare_contact_data(V_test_object)
    
    plt.plot(range(100), np.median(sim_results, axis = 0), 
                     color = color, alpha = 1, label = label)
    plt.fill_between(range(100), np.percentile(sim_results, 95, axis =0), np.percentile(sim_results, 5, axis = 0), 
                     color = color, alpha = 0.3)

    plt.scatter(taniuchi_x_axis, taniuchi_y_axis,marker= 'x', label = 'Taniuchi et al')
    plt.xlabel('Age', fontsize = 15)
    plt.ylabel('Proportion', fontsize = 15)
    plt.title('Age Distribution of HH Contacts', fontsize = 20)
    
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.legend(fontsize = 15)
    plt.savefig(label + '.svg', bbox= 'tight')
    
    
    
    """plt.subplot(1,2,2)
    contact_ages = []
    for infant in infants:
        vid, bid, hhid, iid = infant

        bari = V_test_object.villages[vid].units[bid]
        individual_ages = []
        for hh in bari.households.values():
            individual_ages += [individual.age for individual in hh.individuals.values() if individual.id != iid]

        if len(individual_ages) > 2:
            youngest_two = sorted(individual_ages)[0:2]
        else:
            youngest_two = individual_ages
        contact_ages += youngest_two
    contact_age_counts = Counter(contact_ages)
    x_axis = np.asarray(list(contact_age_counts.keys()))
    y_axis = np.asarray(list(contact_age_counts.values())) / np.sum(list(contact_age_counts.values()))
    plt.bar(x_axis, y_axis, width = 1, color = color, alpha = 0.5, label = label, edgecolor='white')
    plt.xlabel('Age')
    plt.ylabel('Proportion')
    plt.title('Age Distribution of Bari Contacts')
    plt.legend()
"""
```

```python
import scipy.optimize
from scipy.special import gamma as gamma_fn
from scipy.special import gammainc as lower_inc_gamma_fn
from scipy.special import erf


def gamma_pdf(x, params):
    k, theta = params[0], params[1]
    return 1/(gamma_fn(k) * theta**k) * x**(k-1)*np.exp(-x/theta)
def gamma_cdf(x, params):
    k,theta = params[0], params[1]
    return 1/(gamma_fn(k)) * lower_inc_gamma_fn(k, x/theta)


def exponential_pdf(x, params):
    rate = params[0] 
    return rate*np.exp(-rate*x)

def exponential_cdf(x, params):
    rate = params[0]
    return 1-np.exp(-rate*x)

def weibull_pdf(x, params_array):
    rate, k = params_array[0], params_array[1]
    pdf = k/rate * (x/rate)**(k-1) * np.e**-(x/rate)**k
    return pdf

def weibull_cdf(x, params_array):
    rate, k = params_array[0], params_array[1]
    cdf = 1 - np.e**-(x/rate) **k
    return cdf
```

```python
def kl_divergence(pk, qk):
    '''From scipy.stats documentation:
    Calculate the entropy of a distribution for given probability values. (pk vs qk)
    If only probabilities pk are given, the entropy is calculated as S = -sum(pk * log(pk), axis=0).
    If qk is not None, then compute the Kullback-Leibler divergence S = sum(pk * log(pk / qk), axis=0).
    This routine will normalize pk and qk if they dont sum to 1.
    Assumes that pk[i] corresponds to the probablitity of the ith event and that it is a discrete distribution'''
    return scipy.stats.entropy(pk, qk)

def laplace_smoothing(x_array, pseudocount=1):
    N = np.sum(x_array)
    smoothed = (np.asarray(x_array) + pseudocount) / (N + len(x_array)*pseudocount)
    return smoothed

def normalize(array):
    return np.asarray(array) / np.sum(array)

class Param_Search:
    '''Three parameter search'''
    def __init__(self, years=200):
        self.years = 200  
        self.smoothed_bangladesh_dhss_hhsizes = laplace_smoothing([1.9, 10.4, 18.3, 25.2, 20.2, 11.7, 5.8, 3.3, 3.3])
            
    def simulate(self,params):
        hh_move_out_params = params[0:2]
        P = Params(hh_move_out_params = hh_move_out_params, p_stem = 1)
        V= Village.initialize_village(3000, 1)
        for year in range(self.years):
            V.update(1)
        hh_size_count = Counter(V.hh_sizes())
        modified_counts = []
        for number in range(1,9):
            modified_counts.append(hh_size_count.pop(number, 0))
        nine_plus = np.sum(list(hh_size_count.values()))
        modified_counts.append(nine_plus)
        return laplace_smoothing(modified_counts)
    
    def calculate_difference(self, params):
        distance = 0
        simulated_hh_sizes = self.simulate(params)
        distance += kl_divergence(simulated_hh_sizes, self.smoothed_bangladesh_dhss_hhsizes)
        return distance
    
    def calculate_distance_distribution(self, priors):
        '''generates a distance distribution (KL divergence technically) and determines the 30th, 10th, 5th percentiles'''
        distances = []
        for _ in range(100):
            if _ %100 == 0:
                print('calculating distance distribution iteration {iteration} / 1000'.format(iteration = _))
            distance = self.calculate_difference([priors[key]() for key in priors])
            distances.append(distance)
        plt.hist(distances)
        percentiles = [np.percentile(distances, 30), np.percentile(distances, 10), np.percentile(distances, 5), np.percentile(distances, 2.5)]
        return percentiles
    
    def sequential_pmc_ABC(self, n_particles=300, eps = None): 
            '''Population Montecarlo PMC algorithm designed to perform ABC and iteratively adapt the prior/proposal distribution to more closely resemble the posterior
            does this by iteratively reducing the tolerated error and progressively adapting the prior distribution
            follows the Algorithm 2 of Sadegh and vrugt 2014: Approximate Bayesian computation using Markov Chain Monte Carlo simulation: DREAM_ABC_ and the code at https://stackoverflow.com/questions/37496969/population-monte-carlo-implementation
            eps = initial error terms'''
            coefficients_mean = [3.058356842649459, 1.705509538492582]#, 0.8] #4-d polynomial
            scales_std = [1, 0.4]#, 0.05]
            priors = {}
            alphabet_string='abcdefghijklmnopqrstuvwxyz'
            for name, mean, stdev in zip(alphabet_string, coefficients_mean, scales_std):
                priors[name] = lambda: np.random.normal(mean, stdev)


            def computeWeight(prevWeights,prevParticles,prevVariance,currentParticle):
                '''compute the weights needed to adapt the proposal distribution to be closer to the posterior'''
                prior = np.prod([scipy.stats.norm.pdf(currentParticle, coefficients_mean[0], scales_std[0]), 
                                 scipy.stats.norm.pdf(currentParticle, coefficients_mean[1], scales_std[1])]) 
                                 #scipy.stats.norm.pdf(currentParticle, coefficients_mean[2], scales_std[2])])

                denom = 0.0
                scale = 1e5 
                for i in range(len(prevParticles)):
                    denom += prevWeights[i]  * scipy.stats.multivariate_normal(scale * currentParticle, 
                                                                               scale * prevVariance,
                                                                              allow_singular=True).pdf(scale * prevParticles[i])
                return prior/denom    

            if not eps:
                eps = self.calculate_distance_distribution(priors) #automatically calculate the 30th, 10th, 5th percentiles if not given
                print(eps)

            iterations = len(eps)
            distance = 1e9
            variance = {} #for multivariate, this is covariance matrix
            particles = defaultdict(list)
            weights = np.empty([iterations,n_particles])
            simData = np.empty(n_particles)

            for j in range(iterations):
                print('starting {j}th iteration'.format(j=j))
                if j == 0:
                    for i in range(n_particles):
                        while(distance > eps[j]):
                            draw = [priors[key]() for key in priors]
                            distance = self.calculate_difference(draw)
                        #if i in [0, 100, 200, n_particles - 1]:
                        print(draw, distance)
                        print('particle {i} solution found'.format(i = i))
                        particles[j].append(draw)
                        weights[j][i] = 1./n_particles
                        distance = 1e9
                    particles[j] = np.asarray(particles[j])
                    variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
                else:
                    for i in range(n_particles):
                        while(distance > eps[j]):
                            prelim_draw_index = np.random.choice(range(len(particles[j-1])), p = weights[j-1]) #draw a new parameter from the weighted distribution of previously identified parameters; random choice only works on 1-D arrays (no nested)
                            prelim_draw = particles[j-1][prelim_draw_index]
                            #draw = np.random.normal(prelim_draw, np.sqrt(variance[j-1])) #peturb the new prior in proportion to the stdev of the previous iteration
                            #distance = self.calculate_difference([draw])
                            draw = np.random.multivariate_normal(prelim_draw, variance[j-1]) #multivariate_normal accepts the covariance
                            distance = self.calculate_difference(draw)
                        #if i in [0, 100, 200, n_particles - 1]:
                        print('particle {i} solution found'.format(i = i))

                        particles[j].append(draw)
                        weights[j][i] = computeWeight(weights[j-1],particles[j-1],variance[j-1],particles[j][i])
                        distance = 1e9
                    particles[j] = np.asarray(particles[j])
                    variance[j] = 2 * np.cov(particles[j].T) #np.var(particles[j].T)
                    weights[j] = normalize(weights[j])
                print(particles[j])
            self.posteriors = np.asarray(particles)
            self.max_posteriors = [np.mean(distribution) for distribution in self.posteriors.item()[j].T] 
            #at the end of the loop, j is the last iteration

#S=Param_Search()
#eps = [0.016114452661586585, 0.012281604082079253, 0.009966043528806874, 0.008959072890396378]
#S.sequential_pmc_ABC(n_particles = 20)#, #eps = [0.016131239165937644, 0.01207389645355986, 0.010427944785681298]), #0.009066761979215033])
```
