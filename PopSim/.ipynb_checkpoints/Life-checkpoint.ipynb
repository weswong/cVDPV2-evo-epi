```python
class Distribution:
    '''wrapper class for various distributions to optimize'''
    def __init__(self, pdf, cdf, integral=None):
        self.pdf = pdf #or pmf
        self.cdf = cdf
        self.integral = integral  

        
def weibull_distribution(params_array, x):
    rate, k = params_array[0], params_array[1]
    pdf = k/rate * (x/rate)**(k-1) * np.e**-(x/rate)**k
    return pdf

def weibull_cdf(params_array, x):
    rate, k = params_array[0], params_array[1]
    cdf = 1 - np.e**-(x/rate) **k
    return cdf
    
def weibull_integral(params_array, lower, upper):
    integral = weibull_cdf(params_array, upper) - weibull_cdf(params_array, lower)
    return integral

#exponentiated weibull distributions -- 3 parameter distribution, when alpha =1 the same as a weibull
def weibull_exponentiated_distribution(params_array,x):
    rate, k, alpha = params_array[0], params_array[1], params_array[2]
    return alpha * (k/rate)*(x/rate)**(k-1)*(1- np.e**-(x/rate)**k)**(alpha -1) * np.e**(-(x/rate)**k)

def weibull_exponentiated_cdf(params_array, x):
    rate, k, alpha = params_array[0], params_array[1], params_array[2]
    return (1 - np.e **-(x/rate) **k)**alpha

def weibull_exponentiated_integral(params_array, lower, upper):
    integral = weibull_exponentiated_cdf(params_array, upper) - weibull_exponentiated_cdf(params_array, lower)
    return integral

class CMP_distribution:
    '''rough approximation of the conway-maxwell-poisson distribution
    https://en.wikipedia.org/wiki/Conway%E2%80%93Maxwell%E2%80%93Poisson_distribution'''
    def __init__(self):
        pass
    
    @classmethod
    def raw_log_conway_maxwell_poisson(cls, rate, v, x):
        if x != 0:
            log_raw_pdf = (x*np.log(rate)) - v*log_stirling_approximation(x) #unnormalized, logged
        else:
            log_raw_pdf = (x*np.log(rate)) - v*np.log(1) #0! = 1

        return log_raw_pdf
    
    @classmethod
    def pdf(cls, rate, v, x):
        approx_normalizing_constant = np.sum([np.e**CMP.raw_log_conway_maxwell_poisson(rate, v, x) for x in range(100)])
        return (np.e**CMP.raw_log_conway_maxwell_poisson(rate,v, x)) / approx_normalizing_constant
    
    @classmethod
    def cdf(cls, rate, v, x):
        cdf_array = []
        for _ in range(x):
            cdf_array.append(CMP.pdf(rate,v,_))
        return np.sum(cdf_array)
    
    @classmethod
    def plot_example(cls):
        #x is number of events/successes that occur
        plt.title('Conway-Maxwell-Poisson Distribution')
        plt.plot(range(20), [CMP.pdf(1,1.5,x) for x in range(20)])    
        plt.plot(range(20), [CMP.pdf(3,1.1,x) for x in range(20)])
        plt.plot(range(20), [CMP.pdf(5,0.7,x) for x in range(20)])

Poisson = Distribution(lambda params_array, x: scipy.stats.poisson.pmf(x,params_array[0]), #params_array = [lambda], n is a dummy variable
                      lambda params_array, x: scipy.stats.poisson.cdf(x, params_array[0]))
Negative_Binomial = Distribution(lambda params_array, x: scipy.stats.nbinom.pmf(x, params_array[0], params_array[1]),
                                 lambda params_array, x: scipy.stats.nbinom.cdf(x, params_array[0], params_array[1]))
Weibull = Distribution(weibull_distribution, weibull_cdf, weibull_integral) #params_array = rate,k
Weibull_exponentiated = Distribution(weibull_exponentiated_distribution, weibull_exponentiated_cdf, weibull_exponentiated_integral) #params_array = rate, k, alpha
CMP_param = Distribution(lambda params_array, x: CMP_distribution.pdf(params_array[0], params_array[1], x),
                  lambda params_array, x: CMP_distribution.cdf(params_array[0], params_array[1], x))

def sigmoid(p,x):
    x0,y0,c,k=p
    y = c / (1 + np.exp(-k*(x-x0))) + y0
    return y

def multinomial(p_array, x_array):
    n_factorial = gamma_fn(np.sum(x_array) + 1)
    x_i_factorial_product_sum = 1
    for x_i in x_array:
        x_i_factorial = gamma_fn(x_i + 1)
        x_i_factorial_product_sum *= x_i_factorial
    p_i_product_sum = 1
    for x_i, p_i in zip(x_array, p_array):
        p_i_product_sum *= (p_i**x_i)
    
    pmf = (n_factorial / x_i_factorial_product_sum) * p_i_product_sum
    return pmf


```

```python
def read_life_table(file):
    fin = open(file)
    ages = [0]
    males, females = [],[]
    for x in fin.readlines()[1:]:
        age_bins = x.strip().split()[0].split('-')
        if len(age_bins) >1:
            ages.append(np.mean([int(age_bins[0]), int(age_bins[1])]))
        males.append(float(x.strip().split()[1]))
        females.append(float(x.strip().split()[2]))
    ages.append(85.)
    
    return ages, np.asarray(females), np.asarray(males)
```

```python
a_child, b_child = np.polyfit(ages[0:3], np.log(m_mortality[0:3]), deg = 1) #fitting a linear regression model to the log so Y = e**(b + a*x)
a_elder,b_elder = np.polyfit(ages[7:], np.log(m_mortality[7:]), deg = 1)
minimum_mortality =  0.001 #guestimated from the data

def mortality_rate(age, sex, minimum_mortality = 0.001):
    if sex == 1:
        a_child, b_child = np.polyfit(ages[0:3], np.log(m_mortality[0:3]), deg = 1) #fitting a linear regression model to the log so Y = e**(b + a*x)
        a_elder,b_elder = np.polyfit(ages[7:], np.log(m_mortality[7:]), deg = 1)
    elif sex == 0:
        a_child, b_child = np.polyfit(ages[0:3], np.log(f_mortality[0:3]), deg = 1) #fitting a linear regression model to the log so Y = e**(b + a*x)
        a_elder,b_elder = np.polyfit(ages[7:], np.log(f_mortality[7:]), deg = 1)
    #the age thresholds for child_fn and elder_fn are guestimated from the data -- will need to check this when applying to other countries
    #minimum mortality is also guestimated from the data -- this is mostly to do away with a 0 mortality rate
    if age <=12:
        m = max(minimum_mortality, np.e**(b_child + a_child *age))
    elif age > 30:
         m = max(minimum_mortality, np.e**(b_elder + a_elder*age))
    else:
        m = minimum_mortality
    return m
    
```

```python
def read_birth_data(file):
    fin = open(file)
    ages = []
    rates =[] #rates per individual, file reports it as rates per 1000
    age_bins = []
    for x in fin.readlines()[1:]:
        age_bin = x.strip().split()[0].split('-')
        ages.append(np.mean([int(age_bin[0]), int(age_bin[1])]))
        rates.append(float(x.strip().split()[-1]) / 1000.)
        age_bins.append(age_bin)
    return ages, rates, age_bins
```

```python
ages, fertility_rates,raw_age_bins = read_birth_data('data/Matlab_DHSS_age_specific_fertility_2015.txt')

a,b,c = np.polyfit(ages, np.log(fertility_rates), deg = 2) #the relationship between age and log(birthrate) ~ parabolic

def fertility_rate_fn(age, a=-0.012187592249717867, b=0.6277666066187078, c=-9.858673931204683):
    if age < 15 or age > 50:
        return 0
    else:
        return np.e**(a * age**2 + b* age + c)
    
def p_child_fn(age, dt):
    f_rate = fertility_rate_fn(age)
    p_child =1-np.e**(-fertility_rate_fn(age)*dt)
    return p_child 
```

```python
class Individual:
    id_generator = itertools.count()
    scale = 1.0
    def __init__(self, age, base_fertility_params = [1,1,1], scale = 1): #fertility_scale is one of the arguments of the script
        self.id = next(Individual.id_generator) #iterator that generates a new unique id everytime it is called
        self.age = age
        #self.base_fertility_params = base_fertility_params
        self.isalive =True
        self.check_fertility()
        
        
    def check_alive(self, dt): #check if person is still alive after specified time period
        p_death = 1-np.e**(-self.mortality_rate()*dt)
        if np.random.random() < p_death:
            self.isalive = False
            self.death_age = self.age
            
    def check_fertility(self):
        if self.age >=15 and self.age <50: #assume males virility follow fertile rules in females
            self.isfertile = True
        else:
            self.isfertile = False
    
    def polynomial_base_fertility(self):
        # params = [-0.012187592249717867, 0.6277666066187078, -9.858673931204683]
        a,b,c = self.base_fertility_params[0], self.base_fertility_params[1], self.base_fertility_params[2]
        if self.age < 15 or self.age > 50:
            return 0
        else:
            return np.e**(a * self.age**2 + b* self.age + c) *Individual.scale

    def exp_weibull_fertility(self,dt):
        if self.age < 15 or self.age > 50:
            return 0
        else:
            p_child = Weibull_exponentiated.integral(self.base_fertility_params, self.age-15, self.age + dt-15)
            return p_child
    
    
    def base_p_child(self, dt):
        '''baseline probabilities of having a child-- P(Child | no_children)'''
        f_rate = self.polynomial_base_fertility()
        p_child =1-np.e**(-f_rate*dt)
        #p_child = self.exp_weibull_fertility(dt)
        return p_child 

    @classmethod
    def create_child(cls):
        if np.random.random() > 0.5:
            return Male(0)
        else: 
            return Female(0)


class Male(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.sex = 1
    
    def mortality_rate(self, minimum_mortality = 0.001, a = -9.57843000677779, b = 0.08987439511446728):
        #the age thresholds for child_fn and elder_fn are guestimated from the data -- will need to check this when applying to other countries
        #minimum mortality is also guestimated from the data -- this is mostly to do away with a 0 mortality rate
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.038012826363843 + -0.4620757529135555*self.age))
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(a + b*self.age))
        else:
            m = minimum_mortality
        return m
    
    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:
            self.age += dt
            self.check_fertility()
    
    
class Female(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.n_children = 0 #keep track of n_children only on female side
        self.sex =0
        self.t_since_birth = None #period of time where she does not give birth again -- rounding it to 1 year-- this could probably be turned off tbh
        self.birth_age = []

    def mortality_rate(self, minimum_mortality = 0.001, a = -10.527925730119318, b = 0.10104736306154925):
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.154414473758413 + -0.4420609900131315*self.age)) #m = age specific mortality rate (deaths/individual/year)
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(a + b*self.age))
        else: 
            m= minimum_mortality
        return m
    
    def p_birth(self, dt):
        if self.isfertile:
            if self.t_since_birth != None:
                p_birth = self.base_p_child(dt) #* self.birth_interval() * self.birth_desire()
                
            else:
                p_birth = self.base_p_child(dt)
                
        else:
            p_birth = 0
        return p_birth
    
    
    def birth(self, dt):
        '''determines whether a woman gives birth. 
        other than imposing a 1 year refractory period births are 
        treated as independent of one another.. may want to change this'''
        
        if self.isfertile:
            if self.t_since_birth != None:
                p_birth = self.base_p_child(dt) * self.birth_interval() #* self.birth_desire()
                
            else:
                p_birth = self.base_p_child(dt)
    
            if np.random.random() <= p_birth:
                self.n_children +=1
                self.birth_age.append(self.age)
                self.t_since_birth = 0 
                return True
            else:
                return False
        else: 
            return False
    
    def birth_interval(self):
        '''scaling factor that modifies p_child given t years after the previous birth'''
        weibull_exponentiated_params = { '15-19':  [7.008200010283388e-07, 0.18519982325420162, 4955315.919483488],
                                         '20-29': [0.617506311482934, 0.6131814950322227, 15.055481013015964],
                                         '30-39': [0.004432998576956875, 0.25617711748578326, 327.0658166547044],
                                         '40-49': [8.433674445639343e-10, 0.10913540296074727, 110510.2718509799]}
        if self.age >= 15 and self.age <20:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['15-19'], self.t_since_birth)
        elif self.age < 30:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['20-29'], self.t_since_birth)
        elif self.age < 40:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['30-39'], self.t_since_birth)
        elif self.age < 50:
            p_birth_given_time = Weibull_exponentiated.cdf(weibull_exponentiated_params['40-49'], self.t_since_birth)
        else:
            p_birth_given_time = 0
        
        return p_birth_given_time
            
    def birth_desire(self, params = [ 1.3682745,   0.20229474,  0.78551195, -4.10864262]):#obtained from Bangladesh DHSS 2014 Table 6.1
        '''scaling factor that modifies p_child given they have n living children already
        Based on the percent distribution of women aged 15-49 who still desire children given n living children'''
        p_desire = sigmoid(params, self.n_children) #this data is on still LIVING children...
        return p_desire
    
    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:            
            if self.t_since_birth != None:
                self.t_since_birth += dt
            #self.birth(dt)
            self.age += dt #advance time
            #self.check_fertility() #advance time            
```

```python
import scipy.interpolate as interpolate
def inverse_transform_sampling(data, n_bins=20, n_samples=1000):
    hist, bin_edges = np.histogram(data, bins=n_bins, density=True)
    cum_values = np.zeros(bin_edges.shape)
    cum_values[1:] = np.cumsum(hist*np.diff(bin_edges))
    inv_cdf = interpolate.interp1d(cum_values, bin_edges)
    print(list(cum_values),',', list(bin_edges))
    return inv_cdf


test = inverse_transform_sampling([F.death_age for F in f_population])
```

```python
def birth_waits():
    mult_birth_ages = [i.birth_age for i in female_individuals if len(i.birth_age) > 1]
    birth_intervals = []
    for b_times in mult_birth_ages:
        birth_intervals.append([y - x for x,y in zip(b_times,b_times[1:])])

    raw_counts = Counter([_[0] for _ in birth_intervals])
    truncated_counts = defaultdict(lambda: 0) #truncated so that all wait times above 5 accumulated in 5+

    for count, value in raw_counts.items():
        if count <= 5:
            truncated_counts[count] = value / np.sum(list(raw_counts.values()))
        else:
            truncated_counts['5+'] += value / np.sum(list(raw_counts.values()))
        
    return truncated_counts
```

```python
def simulate_n_children(max_age, params):
    female_individuals = [Female(0, params) for _ in range(1000)]
    f_survival = []
    tstep=0
    while tstep < max_age:
        for i in female_individuals:
            i.update(1)
        f_survival.append(sum(total))
        n_children_counts = Counter([i.n_children for i in female_individuals])
        fractionalized_counts  = np.zeros(11)
        for count, value in n_children_counts.items():
            if count < 10:
                fractionalized_counts[count] = value / np.sum(list(n_children_counts.values()))
            else:
                fractionalized_counts[-1] += value / np.sum(list(n_children_counts.values()))
            
        tstep +=1
    return fractionalized_counts

class Children_Number:
    def __init__(self):
        self.raw_data = [_.strip().split(' ') for _ in open('data/Bangladesh_children_distribution_married_women_2014.txt').readlines()]
        self.n_children = {}
        self.mle_params = {}
        for line in self.raw_data[1:-1]:
            age_cohort = line[0]
            n_children_array = [float(x) for x in line[1:12]]
            self.n_children[age_cohort] = n_children_array
            
        self.age_cohorts = list(self.n_children.keys())
            
    def simulate_n_children(self, max_age, params):
        female_individuals = [Female(0, params) for _ in range(400)]
        f_survival = []
        tstep=0
        while tstep < max_age:
            for i in female_individuals:
                i.update(1)
            f_survival.append(sum(total))
            n_children_counts = Counter([i.n_children for i in female_individuals])
            fractionalized_counts  = np.zeros(11)
            for count, value in n_children_counts.items():
                if count < 10:
                    fractionalized_counts[count] = value / np.sum(list(n_children_counts.values()))
                else:
                    fractionalized_counts[-1] += value / np.sum(list(n_children_counts.values()))
            tstep +=1
        return fractionalized_counts
    
    def likelihood(self, params, return_negative = False, randomized = False):
        '''multinomial likelihood'''
        log_p_i_product_sum = 0
        for age_cohort in self.age_cohorts:
            #print(age_cohort)
            x_array = [round(x) for x in self.n_children[age_cohort]]
            p_array = self.simulate_n_children(float(age_cohort.split('-')[-1]), params) 
            tmp = np.asarray(p_array + 0.001)
            regularized = tmp/np.sum(tmp)
            
            x_array = np.asarray(x_array).astype(np.float64)
            p_array = np.asarray(p_array).astype(np.float64)
            for x_i, p_i in zip(x_array, regularized):
                log_p_i_product_sum += (x_i * np.log(p_i/np.sum(p_array)))
        loglikelihood = log_p_i_product_sum
        
        print(params, loglikelihood)
        
        if not return_negative:
            return loglikelihood
        else:
            return -loglikelihood
        
    
    
    def optimize(self, x0=[-0.012187592249717867, 0.6277666066187078, -9.858673931204683]):
        optimize_fn = lambda x0: self.likelihood([x for x in x0], return_negative = True)
        mle_params = scipy.optimize.minimize(optimize_fn, x0, method='BFGS')['x']
        self.mle_params= [x for x in mle_params]
            

    def plot(self, params):
        plt.figure(figsize=(16,10))
        xticks = [x for x in range(0, 10)] + ['10+']
        i = 1
        for age_cohort in self.n_children:
            plt.subplot(3,3,i)
            age_n_children = self.n_children[age_cohort]
            plt.bar(range(len(xticks)), age_n_children/ np.sum(age_n_children), alpha = 0.5, color = 'blue', label = 'Bangladesh DHSS')
            plt.bar(range(len(xticks)), self.simulate_n_children(int(age_cohort.split('-')[-1]), params), color = 'orange', label = 'simulation', alpha = 0.5)
            plt.xticks(range(len(xticks)), xticks)
            plt.title(age_cohort + ' n_children\n per married woman')
            plt.xlabel('Years')
            i+=1
        plt.subplot(3,3,8)
        xes = np.linspace(0,50-15, 1000)
        wxp = [np.e**(params[0] * x**2 + params[1]* x + params[2]) for x in xes]
        plt.plot(xes + 15, wxp, color = 'orange', label = 'new')
        plt.plot(xes + 15,[np.e**(-0.012187592249717867 * x**2 + 0.6277666066187078* x + -9.858673931204683) for x in xes],color ='green', label = 'Original')
        plt.legend()
        plt.tight_layout()
    

N_Children = Children_Number()
```

```python
def multinomial(p_array, x_array):
    n_factorial = gamma_fn(np.sum(x_array) + 1)
    x_i_factorial_product_sum = 1
    for x_i in x_array:
        x_i_factorial = gamma_fn(x_i + 1)
        x_i_factorial_product_sum *= x_i_factorial
    p_i_product_sum = 1
    for x_i, p_i in zip(x_array, p_array):
        p_i_product_sum *= (p_i**x_i)
    
    pmf = (n_factorial / x_i_factorial_product_sum) * p_i_product_sum
    return pmf

def log_stirling_approximation(n):
    '''returns the stirling approximation for factorials
    Stirling approximation is used to approximate the log of a factorial, 
    particularly when n is very large'''
    #return np.sqrt(2*np.pi * n) * (n/np.e)**n
    return n*np.log(n) - n

class Children_Number:
    distributions = {'poisson': Poisson,
                    'cmp': CMP_param}
    nbins = [x for x in range(10)] + ['10+']
    def __init__(self):
        pass
    
    @classmethod
    def initialize(cls, file = 'data/Bangladesh_children_distribution_married_women_2014.txt'):
        C = Children_Number()
        C.n_children_distribution = {} #starting from [0,1,2,3,...10+]
        C.p_children_distribution = {}
        C.total_n = {}
        n_children_data = [_.strip().split(' ') for _ in open('data/Bangladesh_children_distribution_married_women_2014.txt').readlines()][1:-1]
        for line in n_children_data:
            age_group = line[0]
            age_counts = np.asarray([round(float(x)) for x in line[1:12]]) #rounding it because poisson and neg binomial only takes into consideration count data
            C.n_children_distribution[age_group] = age_counts * np.sum(age_counts)
            C.p_children_distribution[age_group] = age_counts / 100.
            
            C.total_n[age_group] = np.sum(age_counts)
        C.age_cohorts = C.n_children_distribution.keys()
        C.mle_params = defaultdict(lambda: {})
        return C
    
    def likelihood(self, distribution_type, params, age_cohort, return_negative = False):
        '''For Poisson, we are trying to estimate lambda (rate)'''
        x_array = self.n_children_distribution[age_cohort] #number of observed successes 
        p_array = []
        
        for nbin in C.nbins[:-1]: 
            p_array.append(Children_Number.distributions[distribution_type].pdf(params, nbin))
        
        p_array.append(1. - Children_Number.distributions[distribution_type].cdf(params, 9)) #10+
        
        log_p_i_product_sum = 0
        for x_i, p_i in zip(x_array, p_array):
            log_p_i_product_sum += (x_i * np.log(p_i))
        loglikelihood = log_p_i_product_sum
        
        if not return_negative:
            return loglikelihood
        else:
            return -loglikelihood
    
    def optimize(self, distribution_type, x0):
        for age_cohort in C.age_cohorts:
            optimize_fn = lambda x0: self.likelihood(distribution_type, [x**2 for x in x0], age_cohort, return_negative = True)
            mle_params = scipy.optimize.minimize(optimize_fn, x0)['x']
            self.mle_params[distribution_type][age_cohort]= [x**2 for x in mle_params]
            
    def plot_data_comparison(self, distribution_type):
        plt.figure(figsize=(16,8))
        for i, age_cohort in enumerate(self.age_cohorts):
            plt.subplot(2,4,i +1)
            
            plt.bar([x for x in range(0,11)], self.p_children_distribution[age_cohort], color = 'blue', label = 'Bangladesh DHSS', width = 0.25)
            
            plt.bar(np.asarray([x for x in range(0,11)]) + 0.25, [Children_Number.distributions[distribution_type].pdf(self.mle_params[distribution_type][age_cohort],x) for x in range(0,11)], color = 'orange', label = 'Poisson', width = 0.25)
            plt.title(age_cohort)
            
C= Children_Number.initialize()
```

```python
class Birth_Intervals:
    distributions = {'weibull':Weibull,
                    'weibull_exponentiated': Weibull_exponentiated}
    
    def __init__(self):
        pass
    
    @classmethod
    def initiate(cls, file='data/Bangladesh_birth_intervals_2014.txt'):
        B = Birth_Intervals()
        B.file = 'data/Bangladesh_birth_intervals_2014.txt'
        B.data = [_.strip().split(' ') for _ in open('data/Bangladesh_birth_intervals_2014.txt').readlines() if '#' not in _][0:-1]
        B.p_age_cohorts = {}
        B.xi_array = {}
        B.n_nonfirst_births = {}
        B.age_cohorts = []
        
        B.birth_intervals = ['7-17', '18-23', '24-35', '36-47', '48-59', '60+']
        B.birth_interval_midpoints = []
        for b in B.birth_intervals:
            if len(b.split('-')) > 1:
                start, stop = float(b.split('-')[0])/12., float(b.split('-')[1])/12.
                B.birth_interval_midpoints.append(np.mean([start, stop]))
        
        B.birth_interval_midpoints.append(5.)
        B.year_limits = [float(x.split('-')[0])/12. for x in B.birth_intervals[:-1]] + [5.]
        
        for line in B.data[1:]:
            B.age_cohorts.append(line[0])
            B.p_age_cohorts[line[0]] = np.asarray([float(x)/100. for x in line[1:7]])
            n_births = float(line[-2].replace(',', ''))
            B.n_nonfirst_births[line[0]] = n_births
            B.xi_array[line[0]] = np.asarray([float(x)/100. for x in line[1:7]]) * n_births 
            
        B.optimize()
        return B
    
    
    def return_randomized(self):
        randomized_dict = defaultdict(list)
        for age_cohort in self.age_cohorts:
            randomized = []
            for p in self.p_age_cohorts[age_cohort]:
                randomized_dict[age_cohort].append(np.random.binomial(B.n_nonfirst_births[age_cohort], p) / B.n_nonfirst_births[age_cohort])
        return randomized_dict
    
    def return_randomized_sample(self, sample_size):
        sampled_dict = defaultdict(list)
        for age_cohort in self.age_cohorts:
            for p in self.p_age_cohorts[age_cohort]: 
                sampled_dict[age_cohort].append(np.random.binomial(sample_size, p))
        return sampled_dict
    
    def multinomial_birth_array_likelihood(self, distribution_type, params, age_cohort, return_negative = False, randomized = False):
        '''assuming wait time distributions follows a weibull or Weibull_exponentiated distribution. 
        DHSS data is discretized into bins, hence the combination of multinomial distributions with the integral range of the Weibull/Weibull_expxonentiated distribution
        The x_array = the x_i for all i year categories where i = {1..5, 5+}'''
        if not randomized:
            x_array = self.xi_array[age_cohort]
        else:
            x_array = self.return_randomized[age_cohort]
            
        
        p_array = []
        for i,start in enumerate(self.year_limits):
            if i != len(self.year_limits) -1:
                stop = self.year_limits[i + 1]
                p_array.append(Birth_Intervals.distributions[distribution_type].integral(params, start, stop))
            else:
                p_array.append(1. - Birth_Intervals.distributions[distribution_type].cdf(params, 5.))

        #not needed...likelihood functions don't need anything that doesn't contain the parameter of interest 
        #(n choose k) stuff are all constants that don't change with p
        #log_n_factorial = log_stirling_approximation(np.sum(x_array))
        #log_x_i_factorial_product_sum = 0
        
        #for x_i in x_array:
        #    log_x_i_factorial = log_stirling_approximation(x_i)
        #    log_x_i_factorial_product_sum += log_x_i_factorial
        
        log_p_i_product_sum = 0
        for x_i, p_i in zip(x_array, p_array):
            log_p_i_product_sum += (x_i * np.log(p_i))
            
        loglikelihood = log_p_i_product_sum #(log_n_factorial - log_x_i_factorial_product_sum) + log_p_i_product_sum
        if not return_negative:
            return loglikelihood
        else:
            return -loglikelihood

        
    def optimize(self):
        self.mle_weibull_params = {}
        self.mle_weibull_exponentiated_params = {}
        for age_cohort in self.age_cohorts:
            optimize_fn = lambda x0: self.multinomial_birth_array_likelihood('weibull', [x**2 for x in x0], age_cohort, return_negative = True) #weibull distributions only defined when all parameters > 0
            mle_params = scipy.optimize.minimize(optimize_fn, [1,1])['x']
            self.mle_weibull_params[age_cohort]= [x**2 for x in mle_params]
            
            optimize_fn = lambda x0: self.multinomial_birth_array_likelihood('weibull_exponentiated', [x**2 for x in x0], age_cohort, return_negative = True)
            mle_params = scipy.optimize.minimize(optimize_fn, [1,1,1])['x']
            self.mle_weibull_exponentiated_params[age_cohort]= [x**2 for x in mle_params]        
            
    
    def plot_mle_distributions(self):
        i=1
        plt.figure(figsize=(8,8))
        for age_cohort in self.age_cohorts:
            plt.subplot(2,2,i)
            plt.plot(range(1,10),[weibull_distribution(self.mle_weibull_params[age_cohort],x) for x in range(1,10)], color = 'orange', label ='weibull')
            plt.plot(range(1,10),[weibull_exponentiated_distribution(self.mle_weibull_exponentiated_params[age_cohort],  x) for x in range(1,10)], color = 'green', label = 'exponentiated weibull')    
            plt.title(age_cohort + ' Birth Intervals')
            plt.legend()
            i += 1
            
    def plot_data_comparison(self):
        plt.figure(figsize=(10,10))
        for i, age_cohort in enumerate(self.age_cohorts):
            plt.subplot(2,2,i +1)
            weibul_sim_values = []
            weibul_exponentiated_sim_values = []
            for x in self.birth_intervals[:-1]:
                age_range = x.split('-')
                if len(age_range) > 1:
                    start, stop = int(age_range[0])/12, (int(age_range[1])+1)/12
                    weibul_sim_values.append(weibull_integral(self.mle_weibull_params[age_cohort], start, stop))

                    weibul_exponentiated_sim_values.append(weibull_exponentiated_integral(self.mle_weibull_exponentiated_params[age_cohort], start, stop))

            weibul_sim_values.append(1.-weibull_cdf(self.mle_weibull_params[age_cohort], 5))
            weibul_exponentiated_sim_values.append(1. - weibull_exponentiated_cdf(self.mle_weibull_exponentiated_params[age_cohort], 5))

            plt.bar([x for x in range(1,7)], self.p_age_cohorts[age_cohort], color = 'blue', label = 'Bangladesh DHSS', width = 0.25)
            plt.bar(np.asarray([x for x in range(1,7)]) + 0.25, weibul_sim_values, color = 'orange', label = 'Weibull', width = 0.25)
            plt.bar(np.asarray([x for x in range(1,7)]) + 0.5, weibul_exponentiated_sim_values, color = 'green', label = 'Exponentiated Weibull', width = 0.25)

            plt.xticks(np.asarray(range(1,7)) + 0.2, self.birth_intervals)
            plt.title(age_cohort + ' Birth Intervals')
        plt.legend()
            
B= Birth_Intervals.initiate()
#runtime errors of zero because it occasionally chooses parameters with distributions like [0,0,0,0,0,1] -- dunno how to fix
```
