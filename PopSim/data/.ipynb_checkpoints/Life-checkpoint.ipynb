```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def read_life_table(file):
    fin = open(file)
    ages = [0]
    males, females = [],[]
    for x in fin.readlines()[1:]:
        age_bins = x.strip().split()[0].split('-')
        if len(age_bins) >1:
            ages.append(np.mean([int(age_bins[0]), int(age_bins[1])]))
        males.append(float(x.strip().split()[1]))
        females.append(float(x.strip().split()[2]))
    ages.append(85.)
    
    return ages, np.asarray(females), np.asarray(males)
```

```python
ages, f_mortality, m_mortality = read_life_table('data/Bangladesh_mortality_rate_per_10000_2015.txt')
```

```python
a_child, b_child = np.polyfit(ages[0:3], np.log(m_mortality[0:3]), deg = 1) #fitting a linear regression model to the log so Y = e**(b + a*x)
a_elder,b_elder = np.polyfit(ages[7:], np.log(m_mortality[7:]), deg = 1)
minimum_mortality =  0.001 #guestimated from the data

def mortality_rate(age, sex, minimum_mortality = 0.001):
    if sex == 1:
        a_child, b_child = np.polyfit(ages[0:3], np.log(m_mortality[0:3]), deg = 1) #fitting a linear regression model to the log so Y = e**(b + a*x)
        a_elder,b_elder = np.polyfit(ages[7:], np.log(m_mortality[7:]), deg = 1)
    elif sex == 0:
        a_child, b_child = np.polyfit(ages[0:3], np.log(f_mortality[0:3]), deg = 1) #fitting a linear regression model to the log so Y = e**(b + a*x)
        a_elder,b_elder = np.polyfit(ages[7:], np.log(f_mortality[7:]), deg = 1)
    #the age thresholds for child_fn and elder_fn are guestimated from the data -- will need to check this when applying to other countries
    #minimum mortality is also guestimated from the data -- this is mostly to do away with a 0 mortality rate
    if age <=12:
        m = max(minimum_mortality, np.e**(b_child + a_child *age))
    elif age > 30:
         m = max(minimum_mortality, np.e**(b_elder + a_elder*age))
    else:
        m = minimum_mortality
    return m
    
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(ages, m_mortality, color = 'DodgerBlue')
plt.scatter(ages, f_mortality, color = 'Violet')
plt.plot(range(85), [mortality_rate(age, 0) for age in range(85)], color = 'Violet')
plt.plot(range(85), [mortality_rate(age, 1) for age in range(85)], color = 'DodgerBlue')
plt.xlabel('Age')
plt.title('Mortality Rate')
plt.ylabel('Mortality Rate/individual/year')

plt.subplot(1,2,2)
plt.scatter(ages, np.log(m_mortality), color = 'DodgerBlue')
plt.scatter(ages, np.log(f_mortality), color = 'Violet')
plt.plot(range(85), np.log([mortality_rate(age, 0) for age in range(85)]), color = 'Violet')
plt.plot(range(85), np.log([mortality_rate(age, 1) for age in range(85)]), color = 'DodgerBlue')
plt.xlabel('Age')
plt.title('Mortality Rate (log)')
plt.ylabel('Mortality Rate/individual/year (log)')
```

```python
def read_birth_data(file):
    fin = open(file)
    ages = []
    rates =[] #rates per individual, file reports it as rates per 1000
    for x in fin.readlines()[1:]:
        age_bins = x.strip().split()[0].split('-')
        ages.append(np.mean([int(age_bins[0]), int(age_bins[1])]))
        rates.append(float(x.strip().split()[-1]) / 1000.)
    return ages, rates
```

```python
ages, fertility_rates = read_birth_data('data/Matlab_DHSS_age_specific_fertility_2015.txt')

a,b,c = np.polyfit(ages, np.log(fertility_rates), deg = 2) #the relationship between age and log(birthrate) ~ parabolic

def fertility_rate_fn(age, a=-0.012187592249717867, b=0.6277666066187078, c=-9.858673931204683):
    if age < 15 or age > 50:
        return 0
    else:
        return np.e**(a * age**2 + b* age + c) 
    
def p_child_fn(age, dt):
    f_rate = fertility_rate_fn(age)
    p_child =1-np.e**(-fertility_rate_fn(age)*dt)
    return p_child 
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.scatter(ages, fertility_rates, color = 'Violet')
plt.xlabel('Age')
plt.ylabel('Births/individual/year')
plt.title('Birth Rates')
plt.plot(range(100), [fertility_rate_fn(_) for _ in range(100)], color = 'Violet')

plt.subplot(1,3,2)
plt.scatter(ages, np.log(fertility_rates), color = 'Violet')
plt.title('Birth Rates (log)')
plt.xlabel('Age')
plt.ylabel('Births/individual/year (log)')
plt.plot(range(100), [np.log(fertility_rate_fn(_)) for _ in range(100)], color = 'Violet')

plt.subplot(1,3,3)
plt.plot(range(100), [p_child_fn(_,1) for _ in range(100)], color = 'Violet')
plt.title('P(Birth)')
plt.xlabel('Age')
plt.ylabel('Probability')

plt.tight_layout()



```

```python
def p_child_fn(age, dt):
    f_rate = fertility_rate_fn(age)
    p_child =1-np.e**(-fertility_rate_fn(age)*dt)
    return p_child 


class Individual:
    id_generator = itertools.count()
    def __init__(self, age):
        self.id = next(Individual.id_generator) #iterator that generates a new unique id everytime it is called
        self.age = age
        self.isalive =True
        self.check_fertility()
        
        
    def check_alive(self, dt): #check if person is still alive after specified time period
        p_death = 1-np.e**(-self.mortality_rate()*dt)
        if np.random.random() < p_death:
            self.isalive = False
            self.death_age = self.age
            
    def check_fertility(self):
        if self.age >=15 and self.age <50: #assume males virility follow fertile rules in females
            self.isfertile = True
        else:
            self.isfertile = False
            
    @classmethod
    def create_child(cls):
        if np.random.random() > 0.5:
            return Male(0)
        else: 
            return Female(0)


class Male(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.sex = 1
    
    def mortality_rate(self, minimum_mortality = 0.001):
        #the age thresholds for child_fn and elder_fn are guestimated from the data -- will need to check this when applying to other countries
        #minimum mortality is also guestimated from the data -- this is mostly to do away with a 0 mortality rate
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.038012826363843 + -0.4620757529135555*self.age))
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(-9.57843000677779 + 0.08987439511446728*self.age))
        else:
            m = minimum_mortality
        return m
    
    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:
            self.age += dt
            self.check_fertility()
    
    
class Female(Individual):
    def __init__(self, *args, **kwargs):
        Individual.__init__(self, *args, **kwargs)
        self.n_children = 0 #keep track of n_children only on female side
        self.sex =0
        self.refractory_period = 0 #period of time where she does not give birth again -- rounding it to 1 year-- this could probably be turned off tbh
    
    def mortality_rate(self, minimum_mortality = 0.001):
        if self.age <=12:
            m = max(minimum_mortality, np.e**(-4.154414473758413 + -0.4420609900131315*self.age)) #m = age specific mortality rate (deaths/individual/year)
        elif self.age > 30:
            m = max(minimum_mortality, np.e**(-10.527925730119318 + 0.10104736306154925*self.age))
        else: 
            m= minimum_mortality
        return m
    
    def birth(self, dt):
        '''determines whether a woman gives birth. 
        other than imposing a 1 year refractory period births are 
        treated as independent of one another.. may want to change this'''
        
        if self.isfertile and self.refractory_period <=0:
            p_birth = p_child_fn(self.age,dt)
            
            if np.random.random() <= p_birth:
                self.n_children +=1
                self.refractory_period = 1 #impose a one year refractory period, but otherwise births
                return True
            else:
                return False
        else: 
            return False
                
    def update(self, dt):
        self.check_alive(dt)
        if self.isalive:            
            self.age += dt #advance time
            self.birth(dt)
            self.check_fertility() #advance time            
            if self.refractory_period > 0:
                self.refractory_period -= dt #going into negative is OK
```

```python
female_individuals = [Female(0) for _ in range(100000)]
male_individuals = [Male(0) for _ in range(100000)]

f_survival = []
total = [i.isalive for i in female_individuals]
while sum(total) > 0:
    for i in female_individuals:
        i.update(1)
    f_survival.append(sum(total))
    total = [i.isalive for i in female_individuals]

m_survival = []
total = [i.isalive for i in male_individuals]
while sum(total) > 0:
    for i in male_individuals:
        i.update(1)
    m_survival.append(sum(total))
    total = [i.isalive for i in male_individuals]

bangladesh_ages, bangladesh_f_survival,  bangladesh_m_survival= read_life_table('data/Bangladesh_survival_2015.txt')
```
