Notice that this Immunoinfection object is not active (is_shed=False) and that it has never been exposed to S2 (naiive=True). This is why the class is referred to as an ImmunoInfection, because it handles both naiive immunity and active infection simultaneously. A special note about IPV_naiive: because IPV recipients don't shed, this flag was put into place to indicate an individual received a  boost in immunity, but I don't think it actually does anything... it's a dangling thread that hasn't been cleaned up yet. 

Infection in this model is dose specific. To give a vaccine-equivalent dose of virus to this ImmunoInfection, run the attempt_infection function:

I.attempt_infection(dose=10**6, modifier = 1)

Dose is the viral dose exposure, by default we assume it is 10**6 for a typical vaccine dose contributes 10**6 virions. Modifier is a setting specific attribute that modulates the efficacy of vaccine. Modifier of 1 means no change. To reduce efficacy, modifier < 1, to increase efficacy, modifier > 1

```python
I.attempt_infection()
print(I.__dict__)
```

Note that the individual is no longer naiive and is now shedding. Also, note that the immunity fields are now filled out. These are OPV-equilent antibody titers (NOT LOG2)!
Again ignore the IPV_naiive attribute 

To advance the infection by one timestep of size dt, run:
    I.update(dt) where dt is expected to be years.
    
I.update(1) advances the infection by one year, I.update(1/365.) will update the infection by a day(ish)

```python
I.update(1/365.)
print(I.__dict__)
```

Example code of how to throw things together and make plots

```python
def reinfection_plots(repetitions, 
                      t_before_challenge = 120, 
                      reinfection_cycles = 5, color = 'blue', alpha = 1, initial_age = 2):
    data = defaultdict(lambda: np.zeros((repetitions, t_before_challenge * reinfection_cycles + 1)))
    for rep in range(repetitions):
        I= ImmunoInfection('S2') # create my ImmunoInfection object using "S2" settings
        #various trackers for plotting
        p_shed = [] 
        p_infection = []
        current_immunity = []
        shed_virions = []
        theta = []
        
        #initial conditions
        p_shed.append(0)  #infection is not active, not shedding
        p_infection.append(1) #P(infection) ~100% given naiive immunity and vaccine dose
        current_immunity.append(0) #not in log2 space, so naiive immunity = 0
        shed_virions.append(0) #not shedding any virus
        theta.append(I.theta_Nab()) #theta is the increase in immunity following succesful exposure
        
        t = 0
        for cycle in range(reinfection_cycles): #how many times are you going to attempt infection
            I.attempt_infection()
            for tstep in range(0,t_before_challenge): #t_before_challenge is the buffer time between infection attempts
                p_shed.append(I.is_shed)
                p_infection.append(I.p_infection())
                current_immunity.append(I.current_immunity)
                current_age = initial_age + t/365.
                shed_virions.append(I.viral_shed(initial_age + (cycle + 1)*(tstep/365.)))
                theta.append(I.theta_Nab())
                t += 1
                I.update(1/365.)

        
        data['p_shed'][rep] = p_shed
        data['p_infection'][rep] = p_infection
        data['current_immunity'][rep] = current_immunity
        data['shed_virions'][rep] = shed_virions
        data['theta'][rep] = np.log2(theta)
    return data
```

```python
t_before_challenge, reinfection_cycles = 720,5
infection_data = reinfection_plots(1000, t_before_challenge, reinfection_cycles)
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Proportion Shedding', y = 1.05, fontsize = 15)
p_shedding = np.mean(infection_data['p_shed'].T, axis = 1)
stdev_shedding = np.sqrt(p_shedding * (1-p_shedding) / 1000)
plt.plot([x for x in range(t_before_challenge*reinfection_cycles + 1)], np.mean(infection_data['p_shed'].T, axis = 1))
plt.fill_between([x for x in range(t_before_challenge*reinfection_cycles+1)], 
                 p_shedding + 2*stdev_shedding,
                 p_shedding - 2*stdev_shedding)
plt.xlabel('Days post initial infection', fontsize = 12)
plt.ylabel('Proportion Shedding', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)


plt.subplot(2,2,3)
plt.title('P(Infection | dose)\n (10^6)', y = 1.05, fontsize = 15)
plt.plot([x for x in range(t_before_challenge*reinfection_cycles+1)], np.median(infection_data['p_infection'], axis = 0))
plt.fill_between([x for x in range(t_before_challenge*reinfection_cycles+1)], 
                  np.percentile(infection_data['p_infection'], 5, axis = 0),
                  np.percentile(infection_data['p_infection'], 95, axis = 0), alpha = 0.3)
plt.xlabel('Days post initial infection', fontsize = 12)
plt.ylabel('Probability', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.subplot(2,2,4)
plt.title('Immunity', y = 1.05, fontsize = 15)
plt.plot([x for x in range(t_before_challenge*reinfection_cycles+1)], np.log2(np.median(infection_data['current_immunity'], axis = 0)))
plt.fill_between([x for x in range(t_before_challenge*reinfection_cycles+1)], 
                  np.log2(np.percentile(infection_data['current_immunity'], 5, axis = 0)),
                  np.log2(np.percentile(infection_data['current_immunity'], 95, axis = 0)), alpha = 0.3)
plt.xlabel('Days post initial infection', fontsize = 12)
plt.ylabel('Log2(Immunity)', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.subplot(2,2,2)
plt.title('Viral Fecal Shed/fecal-oral dose', y = 1.05, fontsize = 15)
plt.plot([x for x in range(t_before_challenge*reinfection_cycles +1)], np.mean(infection_data['shed_virions'], axis = 0))
plt.fill_between([x for x in range(t_before_challenge*reinfection_cycles+1)], 
                  np.percentile(infection_data['shed_virions'], 5, axis = 0),
                  np.percentile(infection_data['shed_virions'], 95, axis = 0), alpha = 0.3)

plt.xlabel('Days post initial infection', fontsize =12)
plt.ylabel('Viral CID50/5ug', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.tight_layout()
```

# Example bOPV reinfection model

bOPV (S1 + S3) administration (reinfection) model for Matlab, Bangladesh. Parameter values chosen to reflect the Taniuchi et al study in the region. The Matlab Model focuses on S2 infection dynamics. bOPV induces a small amount of heterotypic immunity against S2. This code uses an S1 immunoinfection dynamic with a modifier to represent the small amount of cross immunity induced against S2 by the bOPV vaccine. 

```python
def bOPV_reinfection_model():
    #rate is per day!
    I= ImmunoInfection('S1')
    current_immunity, is_shed = [], []                
    for day in range(18*7): #study design lasted for 18 weeks
        I.update(1/365.)
        current_immunity.append(I.current_immunity)
        is_shed.append(I.is_shed)
        if day in 7*np.asarray([6,10,14]): #individuals were exposed to vaccine on week 6, 10, 14
            I.attempt_infection(modifier = 0.31) #modifier to simulate heterotypic S2 immunity
    
    I.attempt_infection()
    for day in range(5*7): #the remaining 5ish weeks after the last vaccine dose
        I.update(1/365.)
        current_immunity.append(I.current_immunity)
        is_shed.append(I.is_shed)
    I.type = ImmunoInfection('S2')
    return np.log2(current_immunity), is_shed #remember! model outputs raw immune values.

#I run this model 1000 times and store the immunity and P(shedding) over time
days = 18*7 +5*7 
bOPV_immunity = np.zeros((1000, days))
bOPV_shedding = np.zeros((1000, days))
for _ in range(1000): 
    bOPV_immunity[_], bOPV_shedding[_] = bOPV_reinfection_model()

#below is just plotting code
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(range(days), np.median(bOPV_immunity, axis = 0), color = 'orange')
plt.fill_between(range(days), np.percentile(bOPV_immunity, 95, axis = 0), np.percentile(bOPV_immunity, 5, axis = 0), 
                 color = 'orange', alpha = 0.3)
plt.scatter([126,126], [np.log2(31), np.log2(31)], marker = 'x', color = 'blue')
plt.title('Simulated bOPV Trial')
plt.xlabel('Time (days)')
plt.ylabel('log2(Nab)')

plt.subplot(1,2,2)
plt.plot(range(days), np.mean(bOPV_shedding, axis = 0), color = 'orange')
plt.scatter([x for x in range(18*7, 24*7, 7)], [0.        , 0.73096447, 0.39593909, 0.34010152, 0.2284264 ,
       0.11167513] , marker = 'x', color = 'blue')
plt.xlim(17*7)
plt.title('Shedding Prevalences post mOPV2 Challenge')
plt.xlabel('Time (days)')
plt.ylabel('Proportion')
```
