```python
import json
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import seaborn as sns
import glob
import re
from collections import Counter
```

```python
class Results:
    def __init__(self, file):
        match = re.search('epidemic_(.+)_(.+)_(.)_(.+).json', file)
        parameters = match.groups()
        self.model = parameters[0]
        self.year = parameters[1]
        self.origin = parameters[2]
        self.p = parameters[3]
        self.results = json.load(open(file))
        self.peak_time = []
        self.peak_size = []
        
        lengths = [len(repetition) for repetition in self.results['total']]
        max_length = np.max(lengths)
        self.max_length = max_length
        for repetition in self.results['total']:
            while len(repetition) != max_length:
                repetition.append(0)
        self.results['fail'] = []
        for repetition in self.results['total']:
            self.results['fail'].append(np.asarray(repetition[1:]) ==0)
        self.results['t_extinction'] = []
        for repetition in self.results['total']:
            peak_idx = np.argmax(repetition)
            self.peak_time.append(peak_idx)
            self.peak_size.append(repetition[peak_idx])
            if 0 in repetition[1:]:
                self.results['t_extinction'].append(repetition[1:].index(0) + 1)
            else:
                self.results['t_extinction'].append(None)
        
    def plot_curves(self, color):
        plt.plot([x for x in range(self.max_length)], 
                 np.mean(self.results['total'], axis = 0), color = color, label = self.year)
        min_range = np.mean(self.results['total'], axis = 0) - 1.96*np.std(self.results['total'], axis = 0)
        min_range =[x if x > 0 else 0 for x in min_range]
        max_range = np.mean(self.results['total'], axis = 0) + 1.96*np.std(self.results['total'], axis = 0)
        plt.fill_between([x for x in range(self.max_length)], min_range, max_range, color = color, alpha = 0.3)
        
    def plot_log_curves(self, color):
        self.log_results = np.log10(self.results['total'])
        plt.plot([x for x in range(self.max_length)], 
                 np.mean(self.log_results['total'], axis = 0), color = color, label = self.year)
        min_range = np.mean(self.log_results['total'], axis = 0) - 1.96*np.std(self.log_results['total'], axis = 0)
        min_range =[x if x > 0 else 0 for x in min_range]
        max_range = np.mean(self.log_results['total'], axis = 0) + 1.96*np.std(self.log_results['total'], axis = 0)
        plt.fill_between([x for x in range(self.max_length)], min_range, max_range, color = color, alpha = 0.3)

        
def read_data(directory):
    multiscale_results, global_results = {}, {}
    c_multiscale_results, c_global_results = {},{}
    for file in glob.glob('C:/Users/wwong/Documents/GitHub/MultiscaleModeling/PopSim/outbreaks/{directory}/epidemic_*json'.format(directory=directory)):
        print(file)
        match = re.search('epidemic_.+_(.+)_._.+.json', file)
        year = int(match.groups()[0])
        match = re.search('epidemic_.+_._(.+).json', file)
        p = float(match.groups()[0])
        print(file)
        param = str(year) + '_' + str(p)
        if p == 0.0:
            if 'multiscale' in file:
                multiscale_results[year]= Results(file)
            else:
                global_results[year] = Results(file)
        else:
            if 'multiscale' in file:
                c_multiscale_results[param] = Results(file)
            else:
                c_global_results[param] = Results(file)
    return multiscale_results, global_results, c_multiscale_results, c_global_results

S2_standard = read_data('standard_S2')
S2_bOPV_standard = read_data('S2_standard_bOPV')

S3_standard = read_data('S3_standard')

WPV_standard = read_data('WPV_standard')
```

```python
def proportioned_histogram(data, bins):
    digitized = np.digitize(data, bins)
    digitized_bin_counts = Counter(digitized)
    total = sum(list(digitized_bin_counts.values()))
    return np.asarray([digitized_bin_counts[k] for k in range(1,42)])/total
```

```python
transmission_multiscale.pkldef p_infection(beta, immunity, dose = 2.5e-06, alpha = 0.44, gamma = 0.46): #dose values are for a typical vaccine dose
        '''Dose response model. Pr(infection | dose, immunity). 
        Immunity is an abstraction of Nab * theta 
        Immunity based on probability of infection'''
        return (1 - (1 + dose/beta)**(-alpha * (immunity)**-gamma)) #* 0.554
```

```python
plt.plot(range(0,12), [p_infection(beta = 2.3, immunity=2**x) for x in range(12)], label = 'WPV')
plt.plot(range(0,12), [p_infection(beta = 8, immunity=2**x) for x in range(12)], label = 'S2')
plt.plot(range(0,12), [p_infection(beta = 18, immunity = 2**x) for x in range(12)], label = 'S3')
plt.legend()

plt.title('P(Infection|2.5e-6 viral dose)')
plt.ylabel('P(Infection)')
plt.xlabel('Immunity (log2Nab)')
```
